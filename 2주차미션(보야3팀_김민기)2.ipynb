📌Q4. 평균 기대수명이 점점 길어지면서 실버 의료 산업도 주목받고 있습니다. 여러분은 실버 의료 산업과 관련된 스타트업에 취업했습니다. 지도를 시각화하여 '노인/치매병원'이 주로 어디에 위치하고 있는지를 찾아보고자 합니다.

folium 을 통해 지도에 전국의 '노인/치매병원'을 표시해 주세요!
다음과 같이 시각화 하며, folium 의 문서를 참고하여 다른 기능을 사용하여 좀 더 멋지게 지도를 꾸며도 좋습니다.
folium 문서 : https://python-visualization.github.io/folium/quickstart.html

```python
import geopandas as gpd
import folium
from folium.plugins import MarkerCluster
import json

#노인병원 추출
df_q4 = df[df["상권업종소분류명"] == "노인/치매병원"]
df_q4

# 지도 기본 위치값
lat = df_q4["위도"].mean()
long = df_q4["경도"].mean()

map = folium.Map(location=(lat, long), zoom_start=9)
#경계선 데이터

#시도별 병원수 data
df_q4_count = df_q4["시도명"].value_counts()
df_q4_1 = pd.DataFrame({'시도명': df_q4_count.index, '병원수': df_q4_count.values})

# JSON 파일 인코딩하여 읽기(시도 경계선 데이터)
with open('/content/Si_Do_map_utf8.json', 'r', encoding='euc-kr') as file:
    Sido_geo = json.load(file)

# 마커 + 뭉치
mc = MarkerCluster()
for n in df_q4.index:
    name = df_q4.loc[n, "상호명"]
    address =str(df_q4.loc[n, "시도명"]) + " " + str(df_q4.loc[n, "시군구명"])
    tooltip = name + "(" + address + ")"
    location = [df_q4.loc[n, "위도"], df_q4.loc[n, "경도"]]
    mc.add_child(folium.Marker(location = location,
                              tooltip=tooltip))

map.add_child(mc)

#밀도 및 경계선 그리기
folium.Choropleth(
    geo_data = Sido_geo, #경계선
    data = df_q4_1, #데이터
    columns=["시도명", "병원수"], #데이터
    key_on = 'feature.properties.CTP_KOR_NM', #경계선
    fill_color="PuRd",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="시도별 노인병원").add_to(map)

map
```

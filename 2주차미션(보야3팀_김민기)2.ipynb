ğŸ“ŒQ4. í‰ê·  ê¸°ëŒ€ìˆ˜ëª…ì´ ì ì  ê¸¸ì–´ì§€ë©´ì„œ ì‹¤ë²„ ì˜ë£Œ ì‚°ì—…ë„ ì£¼ëª©ë°›ê³  ìˆìŠµë‹ˆë‹¤. ì—¬ëŸ¬ë¶„ì€ ì‹¤ë²„ ì˜ë£Œ ì‚°ì—…ê³¼ ê´€ë ¨ëœ ìŠ¤íƒ€íŠ¸ì—…ì— ì·¨ì—…í–ˆìŠµë‹ˆë‹¤. ì§€ë„ë¥¼ ì‹œê°í™”í•˜ì—¬ 'ë…¸ì¸/ì¹˜ë§¤ë³‘ì›'ì´ ì£¼ë¡œ ì–´ë””ì— ìœ„ì¹˜í•˜ê³  ìˆëŠ”ì§€ë¥¼ ì°¾ì•„ë³´ê³ ì í•©ë‹ˆë‹¤.

folium ì„ í†µí•´ ì§€ë„ì— ì „êµ­ì˜ 'ë…¸ì¸/ì¹˜ë§¤ë³‘ì›'ì„ í‘œì‹œí•´ ì£¼ì„¸ìš”!
ë‹¤ìŒê³¼ ê°™ì´ ì‹œê°í™” í•˜ë©°, folium ì˜ ë¬¸ì„œë¥¼ ì°¸ê³ í•˜ì—¬ ë‹¤ë¥¸ ê¸°ëŠ¥ì„ ì‚¬ìš©í•˜ì—¬ ì¢€ ë” ë©‹ì§€ê²Œ ì§€ë„ë¥¼ ê¾¸ë©°ë„ ì¢‹ìŠµë‹ˆë‹¤.
folium ë¬¸ì„œ : https://python-visualization.github.io/folium/quickstart.html

```python
import geopandas as gpd
import folium
from folium.plugins import MarkerCluster
import json

#ë…¸ì¸ë³‘ì› ì¶”ì¶œ
df_q4 = df[df["ìƒê¶Œì—…ì¢…ì†Œë¶„ë¥˜ëª…"] == "ë…¸ì¸/ì¹˜ë§¤ë³‘ì›"]
df_q4

# ì§€ë„ ê¸°ë³¸ ìœ„ì¹˜ê°’
lat = df_q4["ìœ„ë„"].mean()
long = df_q4["ê²½ë„"].mean()

map = folium.Map(location=(lat, long), zoom_start=9)
#ê²½ê³„ì„  ë°ì´í„°

#ì‹œë„ë³„ ë³‘ì›ìˆ˜ data
df_q4_count = df_q4["ì‹œë„ëª…"].value_counts()
df_q4_1 = pd.DataFrame({'ì‹œë„ëª…': df_q4_count.index, 'ë³‘ì›ìˆ˜': df_q4_count.values})

# JSON íŒŒì¼ ì¸ì½”ë”©í•˜ì—¬ ì½ê¸°(ì‹œë„ ê²½ê³„ì„  ë°ì´í„°)
with open('/content/Si_Do_map_utf8.json', 'r', encoding='euc-kr') as file:
    Sido_geo = json.load(file)

# ë§ˆì»¤ + ë­‰ì¹˜
mc = MarkerCluster()
for n in df_q4.index:
    name = df_q4.loc[n, "ìƒí˜¸ëª…"]
    address =str(df_q4.loc[n, "ì‹œë„ëª…"]) + " " + str(df_q4.loc[n, "ì‹œêµ°êµ¬ëª…"])
    tooltip = name + "(" + address + ")"
    location = [df_q4.loc[n, "ìœ„ë„"], df_q4.loc[n, "ê²½ë„"]]
    mc.add_child(folium.Marker(location = location,
                              tooltip=tooltip))

map.add_child(mc)

#ë°€ë„ ë° ê²½ê³„ì„  ê·¸ë¦¬ê¸°
folium.Choropleth(
    geo_data = Sido_geo, #ê²½ê³„ì„ 
    data = df_q4_1, #ë°ì´í„°
    columns=["ì‹œë„ëª…", "ë³‘ì›ìˆ˜"], #ë°ì´í„°
    key_on = 'feature.properties.CTP_KOR_NM', #ê²½ê³„ì„ 
    fill_color="PuRd",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="ì‹œë„ë³„ ë…¸ì¸ë³‘ì›").add_to(map)

map
```

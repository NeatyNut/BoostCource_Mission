✅<2번> #groupby

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df_filtered = df[df['Year'] >= 2011]
df_filtered.groupby(["Year", "Country"])["Life_Expectancy"].mean().unstack()
```

✅<2번> #Pivottable

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df_filtered = df[df['Year'] >= 2011]
df_pivot = df_filtered.pivot_table(index="Year", columns="Country", values="Life_Expectancy", aggfunc="mean")
df_pivot
```
